```python
import pandas as pd
from random import randrange
from datetime import timedelta
from datetime import datetime
import numpy as np
RNG = np.random.default_rng()
```

```python
df1 = pd.read_csv('source_csv/2022-2023 Football Player Stats.csv', sep=';', encoding_errors='ignore')
df1 = df1[df1['Player'].str.contains('\?')==False]
df1
```

```python
def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return str(start + timedelta(seconds=random_second))[:10]

d1 = datetime.strptime('1/1/1980 1:30 PM', '%m/%d/%Y %I:%M %p')
d2 = datetime.strptime('1/1/2010 4:50 AM', '%m/%d/%Y %I:%M %p')
```

```python
players = df1[['Player', 'Rk']].sample(n=500)
players[['name', 'surname']] = players['Player'].str.split(' ', 1, expand=True)
players.rename(columns={'Rk':'player_id'}, inplace=True)
players.drop(columns='Player', inplace=True)
players['birth_date'] = [random_date(d1, d2) for _ in range(players.shape[0])]
players['height'] = [int(RNG.normal(180, 15)) for _ in range(players.shape[0])]
players['weight'] = [int(RNG.normal(85, 15)) for _ in range(players.shape[0])]
```

```python
df2 = pd.read_csv('source_csv/countries of the world.csv', sep=',')
countries = df2[['Country']]
countries.rename(columns={'Country':'name'}, inplace=True)
countries['country_id'] = countries.index
```

```python
players['country_id'] = [int(countries['country_id'].sample(n=1)) for _ in range(players.shape[0])]
```

```python
df3 = pd.read_csv('source_csv/Football teams.csv', sep=',')
leagues = pd.DataFrame()
leagues['name'] = pd.unique(df3['Tournament'])
leagues['league_id'] = leagues.index
```
